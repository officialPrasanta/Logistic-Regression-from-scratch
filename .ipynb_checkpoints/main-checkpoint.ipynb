```python
def sigmoid(z):
    return(1.0/(1 + np.exp(-z)))
```

```python
def predict(features, weights):
    t = np.transpose(weights)
    z = (features.dot(t)).flatten()
    return sigmoid(z)
```

```python
# cost function
# if y_true = 1 then return -log(y_pred)
# else -log(1 - y_pred)
# add each value and divide by observation(length of y_true)

def cross_entropy(y_true, y_pred):
    cost = np.array([(-1*(i*np.log(j) + (1-i)*np.log(j))) for i, j in np.nditer([y_true.flatten(), y_pred])])
    return (cost.sum()/len(y_true))
```

```python
def get_gradient(features, y_true, y_pred, lr):
    gradient = np.dot(np.transpose(features), (y_true - y_pred))*lr*(1.0/len(y_true))
    return gradient
```

```python
def LogisticRegression(features, y_true, epochs, lr=1e-5):
    weights = np.random.rand(features.shape[1])
    for i in range(epochs):
        y_hat = predict(features, weights)
        cost = cross_entropy(y_true, y_hat)
        w_grad = get_gradient(features, y_true, y_hat, lr)
        weights -= w_grad
        if(i % 100 == 0):
            print('Iteration: {}, Loss: {}'.format(i, cost))
    
    y_pred = [(0, 1)[i>0.5] for i in y_hat]
    return y_pred
```
